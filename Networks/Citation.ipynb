```python
import pandas as pd
from habanero import Crossref

# Cargar el CSV con tus artículos
df = pd.read_csv('D:\\CodeProjects\\QuantFinanceResearch\\CSVs\\AssetPricingModels.csv')

# Eliminar artículos que tengan DOI nulo
df_cleaned = df.dropna(subset=['DOI'])

# Mantener solo las columnas necesarias
df_reduced = df_cleaned[['DOI', 'Title', 'Year', 'Source title', 'Page count']]

# Ordenar el dataframe por la columna 'Page count' para encontrar los más citados
df_most_cited = df_reduced.sort_values(by='Page count', ascending=False).head(1000)

# Función para obtener las referencias de cada DOI
def get_references(doi):
    cr = Crossref()
    try:
        result = cr.works(ids=doi)
        references = result['message'].get('reference', [])
        return [ref.get('DOI') for ref in references if ref.get('DOI')]
    except:
        return []

# Filtrar artículos que tienen referencias disponibles
df_most_cited['Cited_DOIs'] = df_most_cited['DOI'].apply(get_references)
df_most_cited_filtered = df_most_cited[df_most_cited['Cited_DOIs'].map(len) > 0]

# Guardar solo los artículos que tienen citas disponibles
df_most_cited_filtered.to_csv('articulos_mas_citados_con_referencias.csv', index=False)

# Imprimir muestra del nuevo dataframe
print(df_most_cited_filtered.head(50))
```
